# 1. Pyplot API (Simple but more effective for beginner)

### Univariate Analysis - Numerical Data

```python
### GEnerating of dataframe using Gemini
import pandas as pd
import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

# Define the number of rows
num_employees = 50

# 1. Department Feature: Randomly select from the list (HR, IT, Finance)
departments = np.random.choice(['HR', 'IT', 'Finance'], size=num_employees)

# 2. Salary Feature: Generate random salaries (e.g., between $40,000 and $150,000)
# Round to the nearest 100
salaries = np.round(np.random.uniform(40000, 150000, size=num_employees) / 100) * 100

# 3. Age Feature: Generate random ages (e.g., between 22 and 60)
ages = np.random.randint(22, 61, size=num_employees)
experiance = np.random.randint(1,5,size=num_employees)
# Create the DataFrame
df = pd.DataFrame({
    'department': departments,
    'salary': salaries.astype(int), # Convert to integer
    'age': ages,
    'experiance':experiance
})

# Display the first few rows and the shape
print("--- Generated DataFrame (First 5 Rows) ---")
print(df.head())
print("\n--- DataFrame Shape ---")
print(df.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.show()
```

```python
## 1. Line plot
df['salary'] = np.sort(df['salary'])
plt.plot(df['salary'],marker='o',linestyle='--',linewidth=1,color='red')
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python
### 2. Histogram
plt.hist(df['salary'],bins=30,color='green',align='mid')
plt.xlabel("Frequency")
plt.ylabel("Count")
plt.title("Histogram plot")
plt.show()
```

```python
### 3. Boxplot
plt.boxplot(df['salary'])
plt.ylabel("Salary")
plt.title("Box Plot")
plt.show()
```

### Univariate Analysis - Categorical Data

```python
### 1. Pie chart
count = df['department'].value_counts()
plt.pie(count,labels=count.index,autopct="%1.1f",explode=[0.0,0.1,0.0],shadow=True)
plt.show()
```

```python
### 2. CountPlot
plt.bar(count.index,count,label=['Finance','Hr','IT'],color=['red','green','pink'])
plt.legend()
plt.xlabel("Department")
plt.ylabel("Count")
plt.title("CountPlot")
plt.show()
```

## Bivariate Analysis - Numerical data


```python
### 1. Scatter plot (numerical-numerical)
plt.scatter(x=df['salary'],y=df['age'],color='red')
plt.xlabel("salary")
plt.ylabel("age")
plt.title("Scatterplot-Bivariate")
plt.show()
```

```python
### 2. Lineplot
plt.plot(df['salary'],df['age'],marker='o',linestyle='--',color='red')
plt.xlabel("salary")
plt.ylabel("age")
plt.title("LinePlot-Bivariate")
plt.legend()
plt.show()
```

```python
### 3. Bar Chart
plt.bar(df['salary'],df['age'])
plt.xlabel("salary")
plt.ylabel("age")
plt.title("BarPlot-Bivariate")
plt.show()
```

## bivariate - (numerical - categorical )

```python
hr_salary = df[df['department']=='HR']['salary']
it_salary = df[df['department']=='IT']['salary']
finance_salary = df[df['department']=='Finance']['salary']
### 1. Boxplot
plt.boxplot([hr_salary,it_salary,finance_salary],labels=['HR','IT','Finance'])
plt.xlabel("Features")
plt.legend()
plt.show()
```

```python
### 2. Pie chart
count = df.groupby('department')['salary'].mean()
plt.pie(count,labels=count.index,autopct="%1.1f",explode=[0.0,0.1,0.0],shadow=True)
plt.show()
```

```python
### 3. Barplot
plt.bar(count.index,count)
plt.show()
```

## Multivariate Analysis (numerical - numerical - numerical)


```python
### 1. bubble plot
### = here the third column represent the width of the bubble(discrete variable)
plt.scatter(df['salary'],df['age'],s=df['experiance']*20)
plt.show()
```

### Multivariate Analysis - (2 numerical and 1 categorical)

```python
color = {'HR':'yellow','Finance':'red','IT':'green'}
copy_department = df['department'].copy()
plt.scatter(df['salary'],df['age'],c=copy_department.map(color))
plt.show()
```

## 3D - plot


```python
### Using matplot
hr_salary = df[df['department']=='HR']['salary']
it_salary = df[df['department']=='IT']['salary']
finance_salary = df[df['department']=='Finance']['salary'][:-5]
ax = plt.axes(projection='3d')
ax.scatter(hr_salary,it_salary,finance_salary)
plt.show()
```
